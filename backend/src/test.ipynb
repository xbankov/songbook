```python
text_lines = html_text.replace("\r", "").split("\n")


def is_ug_tag(text: str) -> bool:
    stripped = text.strip()
    return (
        "\n" not in stripped
        and stripped.startswith("[")
        and stripped.endswith("]")
        and "[tab]" not in stripped
        and "[ch]" not in stripped
    )


current_section_start = 0
for i, line in enumerate(text_lines):
    # Start of the section
    if is_ug_tag(line):
        if i > current_section_start:
            current_section_end = i
            print(f"{current_section_start} - {i}")
            for line in text_lines[current_section_start:current_section_end]:
                print(line)
            current_section_start = i
    # Last section
    elif i == len(text_lines) - 1:
        current_section_end = i
        print(f"{current_section_start} - {i}")
        for line in text_lines[current_section_start : i + 1]:
            print(line)
            # section_lines = text_lines[current_section_start:current_section_end]
            # parsed_sections.append(Section.from_chordpro("\n".join(section_lines)))
```
